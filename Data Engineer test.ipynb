#### b. For Bogota, what neighborhoods reported greater amount of murders between 6 at night and 12 midnight?

```python
#Group data by municipality Bogotá
Bogota= data.set_index('Municipio')
Bogota=Bogota[Bogota.index=='BOGOTÁ D.C. (CT)']
#Changing the data type of column Hora into data time type 
horas=pd.to_datetime(Bogota['Hora']).dt.hour.values
Bogota['Hora']=horas
#Selecting all the murders that ocurr between 6 an 24 hours 
Bogota_hours=Bogota[(Bogota['Hora']>6) & (Bogota['Hora']<24)]
#Group the data by neighborhood an sum the quantity of murders
neigh_murder=Bogota_hours.groupby(['Barrio'])[['Cantidad']].agg('sum')
#soring out in  descending order and selecting only the first 10 
top10_neigh=neigh_murder.sort_values('Cantidad',ascending=False)[:10]
#Plotting the data 
x_neigh=top10_neigh['Cantidad'].keys()
y_neigh=top10_neigh['Cantidad'].values
plt.xticks(rotation='vertical')
plt.bar(x_neigh,y_neigh)
plt.xlabel('Neighborhood')
plt.ylabel('Number of  murders')
plt.show()
```

As we can see from the graph above  the neighborhoods with the higher number of murders between the 6 at night and midnight are: la favorita with 9 murders, los alpes, las margaritas, gran britalia, diana turbay, arborizadora alta with 4 murders each one and barranquillita, madelena and la estanzuela with 3.

```python
#Select only the two categories 
data_set_tw=dataset[(dataset['Arma_empleada']=='ARMA BLANCA / CORTOPUNZANTE') | (dataset['Arma_empleada']=='ARMA DE FUEGO')]
```

As we can see we have different categorical variables like day, zone, place category and mobilities so it's necessary to make a code format for each dimension. In this case I am going to use ordinal encoders from sklearn to make the codification.   I think it is the best way of encoding in this case because there are some variables like place class that have 111 different types and other encoder like hot encoding could increase the dimension of the problem

```python
#Zona encoder 
from sklearn.preprocessing import OrdinalEncoder
z_encode=OrdinalEncoder()
categories_z=[['URBANA'],['RURAL']]
z_encode.fit(categories_z)
zona_column=data_set_tw['Zona']
zone_transformed=z_encode.transform(zona_column.values.reshape(-1,1))
data_set_tw['Zona']=zone_transformed
```

```python
#clase_de_Sitio encoder
cs_encode=OrdinalEncoder()
categories_cs=data_set_tw.groupby(['Clase_de_sitio'])['Zona'].max() #grouping by clase de sitio
categories_cs=categories_cs.keys().to_numpy()
cs_encode.fit(categories_cs.reshape(-1,1))
cs_column=data_set_tw['Clase_de_sitio']
cs_transformed=cs_encode.transform(cs_column.values.reshape(-1,1))
data_set_tw['Clase_de_sitio']=cs_transformed
```

```python
#day encoder 
day_encode=OrdinalEncoder()
categories_day=data_set_tw.groupby(['DIa'])['Zona'].max() #grouping by day
categories_day=categories_day.keys().to_numpy()
day_encode.fit(categories_day.reshape(-1,1))
day_column=data_set_tw['DIa']
day_transformed=day_encode.transform(day_column.values.reshape(-1,1))
data_set_tw['DIa']=day_transformed
```

```python
#mobility assesin encoder 
mb_encode=OrdinalEncoder()
categories_mb=data_set_tw.groupby(['Movil_Victima'])['Zona'].max() #grouping by mobility 
categories_mb=categories_mb.keys().to_numpy()
mb_encode.fit(categories_mb.reshape(-1,1))
mb_a_column=data_set_tw['Movil_Agresor']
mb_a_transformed=mb_encode.transform(mb_a_column.values.reshape(-1,1))
data_set_tw['Movil_Agresor']=mb_a_transformed
```

```python
#The same encoder  from assesin mobility can be used to victim mobility
mb_v_column=data_set_tw['Movil_Victima']
mb_v_transformed=mb_encode.transform(mb_v_column.values.reshape(-1,1))
data_set_tw['Movil_Victima']=mb_v_transformed
```

```python
#Sex encoder
sx_encode=OrdinalEncoder()
categories_sx=data_set_tw.groupby(['Sexo'])['Zona'].max() #grouping by mobility 
categories_sx=categories_sx.keys().to_numpy()
sx_encode.fit(categories_sx.reshape(-1,1))
sx_column=data_set_tw['Sexo']
sx_transformed=sx_encode.transform(sx_column.values.reshape(-1,1))
data_set_tw['Sexo']=sx_transformed
```

```python
#print the current data set
data_set_tw
```

As we can see above there are not categorical columns only our output of the model (Arma_empleada)is still categorial. We're going to deal with this in a moment. However the data related to dates and time is a string. We need to make something to convert it into a number variable with the objective that at the training moment of some model there won't be errors. So what I'm going to do is separate the date in two new variables (day, month ) and the time in two new Hour and minute. The year of the date is not take into acccount because is the same for all the data. 

```python
#Two new variables for the time 
Horas=pd.to_datetime(data_set_tw['Hora']).dt.hour.values #Hours
Minutos=pd.to_datetime(data_set_tw['Hora']).dt.minute.values #minutes
#Add the new variables 
data_set_tw['Horas']=Horas
data_set_tw['Minutos']=Minutos
#Drop the old variable of time 
data_set_tw=data_set_tw.drop(columns=['Hora'])
```

```python
#Three new variables for the date 
dia=pd.to_datetime(data_set_tw['Fecha_corta']).dt.day.values #day
mes=pd.to_datetime(data_set_tw['Fecha_corta']).dt.month.values #month

#Add the new variables 
data_set_tw['Dias']=Horas
data_set_tw['Meses']=Minutos
#Drop the old variable of time 
data_set_tw=data_set_tw.drop(columns=['Fecha_corta','Conteo'])
```

With all the variables pre processed, the next step is create a the output vector that will be the labels. 

```python

# Encoder for the output 
output_encode=OrdinalEncoder()
categories_ou=data_set_tw.groupby(['Arma_empleada'])['Zona'].max() #grouping by mobility 
categories_ou=categories_ou.keys().to_numpy()
output_encode.fit(categories_ou.reshape(-1,1))
ou_column=data_set_tw['Arma_empleada']
ou_transformed=output_encode.transform(ou_column.values.reshape(-1,1))
data_set_tw['Arma_empleada']=ou_transformed
Y=data_set_tw['Arma_empleada'].values #new vector 
#Delete the output from the input variables 
data_set_tw=data_set_tw.drop(columns=['Arma_empleada'])
#Final_data_set
final_data_set=data_set_tw.to_numpy()
final_data_set
```

Now we have all input variables in numeric format and we have create the label vector with numeric format. The next step is to split the data set into the training and evaluation data sets. I'm going to use 80% to train the model. With this 80% I'm going to choose the 20% for measure the model accuaracy and try to improve the model varying the parameters and when I decided which is the final model I'm going to report with this the final metrics of performance. 

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train_complete,X_test,y_train_complete,y_test=train_test_split(final_data_set,Y,test_size=0.2) #test and training set
#From the training data set take the 20% for the validation
X_train,X_validation,y_train,y_validation=train_test_split(X_train_complete,y_train_complete,test_size=0.2)
```

To begin to choose a model first I'm going to decide which kind of model. Like this is a classification problem I'm going to probe by default parameters which is the most suitable between random forest, support vector machine and neural networks, after I'm going to do a search of parameters. To choose which is the best model I'm going to use the F1 score measure. Because this measure implies precision and recall for the model

```python
#Random forest classificator
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
random_f = RandomForestClassifier(max_depth=2, random_state=0)
random_f.fit(X_train, y_train)
y_random_forest=random_f.predict(X_validation)
f1_rf=f1_score(y_validation,y_random_forest)
```

```python
#Support vector machine 
#For the SVC the data must be normalized so I'll do a Standard scaler 
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
X_train_scal=scaler.transform(X_train) #training data scaled
SuppVec=SVC(gamma='auto') #Defining parameters for SVC 
SuppVec.fit(X_train_scal,y_train)
X_val_scal=scaler.transform(X_validation)
y_SVM=SuppVec.predict(X_val_scal)
f1_SVM=f1_score(y_validation,y_SVM)
```

```python
#Neural network classifier
from sklearn.neural_network import MLPClassifier
NeuralNetwork=MLPClassifier(random_state=1,max_iter=600)
NeuralNetwork.fit(X_train_scal,y_train)
y_MLP=NeuralNetwork.predict(X_val_scal)
f1_MLP=f1_score(y_validation,y_MLP)
```

```python
f1_vals=[f1_rf,f1_SVM,f1_MLP]
model=['Random Forest', 'SVM', 'Neural Network']
```

```python
plt.bar(model,f1_vals)
plt.xlabel('Model')
plt.ylabel('F1-score')
plt.title('F1-score for each model')
plt.show()
```

From the graph above it is possible to see that the three model has similar F1-score. But the best is for the Random Forest. To try to improve the model,I will make a search on some parameters of the model.  The first parameter is the deph of the random forest:

```python
classifiers_rf=[ RandomForestClassifier(n_estimators=100, max_depth=i) for i in range(2,50,3)] #creating classifiers

rf=[i for i in range(2,100,2)]
score_rf_val=np.zeros((len(classifiers_rf),))
score_rf_test=np.zeros((len(classifiers_rf),))
#Loop to  train and evaluate differnt random forest with different depth 
for i in range(len(classifiers_rf)):
    classifiers_rf[i].fit(X_train, y_train) #train the classifier
    y_pred=classifiers_rf[i].predict(X_validation)  #Predict with x validation
    score_rf_val[i]=f1_score(y_validation,y_pred)
    y_pred_t=classifiers_rf[i].predict(X_test)  #Predict with x validation
    score_rf_test[i]=f1_score(y_test,y_pred_t)
```

```python
#Plot the f1 measure 
plt.plot(range(2,50,3),score_rf_val)
plt.title('F1 score for differen depths of Random Forest')
plt.xlabel('Depth')
plt.ylabel('F1 score')

print('The greates F1 score is ',np.max((score_rf_val)))
```

So the final model chosen is a random forest with a F1 score of 0.8833. Which parameters are: 
number of trees (n_estimator) =100
maximum depth of the trees= 5
minimum number of samples =2
